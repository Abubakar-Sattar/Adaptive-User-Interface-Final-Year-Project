## Handling Missing Data

```python
# Handling Missing Data

# Shape of Our Dataset

dataset.shape

# Checking Missing Values
dataset.isnull()

# Number of Missing Values
dataset.isnull().sum()

# Total Number of Missing Values
dataset.isnull().sum().sum()

# Filling Missing Values

# List of columns with missing values
columns_with_missing_values = ['Country', 'Gender', 'Location', 'City', 'Shopping Habit',
                               'Device for Shopping(Smart phone)', 'Device for Shopping(Desktop and Laptop)',
                               'Products(Clothing)', 'Products(Electronics)', 'Products(Other)',
                               'Source Of Discvorey', 'Influencing Factors', 'Required Categories',
                               'Design Factors', 'Color Scheme', 'Age']  

for column in columns_with_missing_values:
    # Compute the mode for each column
    mode_value = dataset[column].mode()[0]
    
    # Fill the missing values with mode for each column
    dataset[column].fillna(mode_value, inplace=True)

# Printing Dataset

dataset

```

## Outlier Detection

```python
#Outlier Detection

# List of columns with categorical data,
categorical_columns = [
    'Gender', 'Location', 'Country', 'City', 'Shopping Habit', 'Device for Shopping(Smart phone)',
    'Device for Shopping(Desktop and Laptop)', 'Products(Clothing)', 'Products(Electronics)',
    'Products(Other)', 'Source Of Discvorey', 'Influencing Factors', 'Required Categories',
    'Design Factors', 'Color Scheme', 'Age'
]

# Define a threshold for outlier detection (adjust as needed)
threshold = 10  # For example, considering categories with a count less than 10 as outliers

for column in categorical_columns:
    # Calculate frequency of each category
    category_counts = dataset[column].value_counts()
    
    # Identify outliers (categories with counts below the threshold)
    outliers = category_counts[category_counts < threshold]
    
    # Print or handle the outliers
    print(f"Outliers in '{column}':")
    print(outliers)
    print()
```

## Outlier Treatment

```python
# Outliers Treatment

# List of all columns
all_columns = dataset.columns.tolist()

# Define a threshold for outlier detection based on category frequency
threshold = 10  # For example, considering categories occurring less than 10 times as outliers

# Handle outliers in all columns by labeling rare categories as 'Other'
for column in all_columns:
    category_counts = dataset[column].value_counts()
    outliers = category_counts[category_counts < threshold].index.tolist()
    dataset[column] = dataset[column].apply(lambda x: 'Other' if x in outliers else x)


```

## Hot Encoding

```python
# Hot Encoding

# Using Sklearn

from sklearn.preprocessing import OneHotEncoder

# Convert all object-type columns to categorical columns
for col in dataset.select_dtypes(include=['object']).columns:
    dataset[col] = dataset[col].astype('category')

# Initialize OneHotEncoder
encoder = OneHotEncoder()

# Fit and transform all columns
encoded_data = encoder.fit_transform(dataset)

# Convert the encoded data to a DataFrame
encoded_dataset = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out(dataset.columns))
encoded_dataset
```

```python
# Create age bins based on the encoded columns
age_bins = [0, 1, 2, 3, 4, 5]  # Example bins for the encoded age columns
age_labels = ['Under 18', '18-24', '25-34', '35-44', 'Other']  # Labels for age bins

# Get age category for each row
encoded_dataset['Age_Category'] = encoded_dataset[['Age_Under 18', 'Age_18-24', 'Age_25-34', 'Age_35-44', 'Age_Other']].idxmax(axis=1)
encoded_dataset['Age_Category'] = encoded_dataset['Age_Category'].map({
    'Age_Under 18': 'Under 18',
    'Age_18-24': '18-24',
    'Age_25-34': '25-34',
    'Age_35-44': '35-44',
    'Age_Other': 'Other'
})

# Initialize 'Gender_Category' column
encoded_dataset['Gender_Category'] = 'Other'

# Update 'Gender_Category' based on age categories
encoded_dataset.loc[encoded_dataset['Age_Category'] == 'Under 18', 'Gender_Category'] = 'Female'
encoded_dataset.loc[encoded_dataset['Age_Category'] != 'Under 18', 'Gender_Category'] = 'Other'  # For other categories

# For non-'Under 18' rows, determine the gender based on the maximum value between 'Gender_Female' and 'Gender_Male'
gender_columns = ['Gender_Female', 'Gender_Male']
mask = encoded_dataset['Age_Category'] != 'Under 18'
encoded_dataset.loc[mask, 'Gender_Category'] = encoded_dataset.loc[mask, gender_columns].idxmax(axis=1).map(lambda x: x.replace('Gender_', ''))
```

## Clustering

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Selecting columns for clustering (considering all columns except Age and Gender)
clustering_data = encoded_dataset.drop(['Age_Category', 'Gender_Category'], axis=1)

# Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(clustering_data)

# Number of clusters
num_clusters = 7

# KMeans clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
encoded_dataset['Cluster'] = kmeans.fit_predict(scaled_data)

# For individuals under 18, create clusters based on everything present in other columns
encoded_dataset.loc[encoded_dataset['Age_Category'] == 'Under 18', 'Cluster'] = encoded_dataset[encoded_dataset['Age_Category'] == 'Under 18'].drop(['Age_Category', 'Gender_Category', 'Cluster'], axis=1).apply(tuple, axis=1).map(hash) % num_clusters
```
